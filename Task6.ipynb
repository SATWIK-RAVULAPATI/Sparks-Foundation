**Importing Libraries**


```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
```

**Loading DataSet**

```python
data_frame=pd.read_csv(r'Iris.csv')
```

**Top 10 Rows of the DataFrame**

```python
data_frame.head(10)
```

```python
data_frame.shape
```

***Checking for any null value:***

```python
data_frame.info()
```

```python
data_frame.describe()
```

**Setting Index as Id**

```python
data_frame=data_frame.set_index("Id")
```

```python
data_frame.head(10)
```

**Spliting the data for Training and Testing**

```python
import sklearn
from sklearn.model_selection import train_test_split
```

**X as Input Variable**

```python
X=data_frame.iloc[:,0:4].values
```

```python
X
```

**Y as target Variable**

```python
Y=data_frame.Species
```

```python
Y
```

```python
X.shape
```

**20% data is used for testing and 80% data will used for training**

```python
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size = 0.2, random_state = 100)
```

**importing DescisionTreeClassifier**

```python
from sklearn.tree import DecisionTreeClassifier 
```

```python
d_tree=DecisionTreeClassifier(random_state=100)
```

**Fitting the TrainingData**

```python
d_tree.fit(X_train,y_train)
```

**Predicting the Species of TestData**

```python
predictions=d_tree.predict(X_test)
```

```python
predictions
```

**Inserting Random data To predict the Species**

```python
predict=d_tree.predict([[6.7,5.6,7.1,5.3]])
predict
```

**Calculating The Accuracy of the Model,Confusion Matrix and Cassification Report**

```python
from sklearn.metrics import accuracy_score 
from sklearn.metrics import confusion_matrix,classification_report 
```

```python
print("classification report of the model:\n",classification_report(y_test,predictions))
```

```python
print("Confusion_matrix:\n",confusion_matrix(y_test,predictions))
```

```python
print("Accuracy of the model:",accuracy_score(y_test,predictions))
```

   **Accuracy of our model is 96.66%**
